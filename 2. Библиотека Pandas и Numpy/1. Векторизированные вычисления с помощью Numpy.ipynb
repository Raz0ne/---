### Хранение массивов на диске в двоичном формате

_np.save_ и _np.load_ - основное функции для эффективного сохранения и загрузки данных с диска. По умолчанию массивы хранятся в несжатом двоичном формате в файле с расширением _.npy_.

```python
arr = np.arange(10)
```

```python
np.save('some_array', arr)
```

```python
np.load('some_array.npy')
```

Можно сохранить несколько массивов в zip-архиве с помощью функции np.savez, которой массивы передаются в виде именнованных аргументов.

```python
np.savez('array_archive.npz', a=arr, b=arr)
```

```python
arch = np.load('array_archive.npz')
```

```python
arch['b']
```

### 2) Выведите средний размер пополнений за февраль подневно, в разбивке по валютам.

### Результатом должна быть таблица формата "дата, валюта и размер пополнения" (с округлением до целого числа).

```python
sql2 = """
SELECT 
    DATE(operation_date) AS day,
    currency,
    CAST(ROUND(AVG(volume)) AS Integer) AS mean_volume
FROM 
    income
GROUP BY 
    day,
    currency
"""
```

## 3) Посчитайте долю клиентов, у которых были пополнения в нескольких валютах. Ответ округлите до десятых процента.

```python
sql3 = """
WITH t1 AS
(
SELECT
    user_id
FROM
    income
GROUP BY
    user_id
HAVING
    COUNT(DISTINCT currency) > 1
),
t2 AS 
(
SELECT
    DISTINCT user_id
FROM
    income
),
t3 AS 
(
SELECT
    CASE
        WHEN
            t1.user_id IS NOT NULL
            THEN 1
        ELSE 0
    END as presence
FROM
    t2
LEFT JOIN
    t1
ON
    t1.user_id = t2.user_id
)
SELECT
    ROUND(SUM(presence) * 1.0 / COUNT(*) * 100, 1) as share
FROM 
    t3
"""
```

```python
pd.read_sql(sql3, connection)
```

## 4) По каждому клиенту вывести дату первого пополнения, валюту, сумму в рублях (используйте вторую таблицу). Ответом выведите общую сумму в рублях по всем клиентам.

```python
sql4 = """
WITH t AS
(
SELECT
    rate_currency,
    rate_countercurrency,
    CAST(CASE
        WHEN 
            LENGTH(rate_value) > 10
            THEN SUBSTR(rate_value, 6, 2) || "." || SUBSTR(rate_value, 1, 4)
        ELSE
            rate_value
    END AS REAL) as rate_value,
    rate_date
FROM
    currencies
)
SELECT
    user_id,
    currency,
    operation_date,
    volume * CASE
        WHEN
            t.rate_value IS NULL
            THEN 1.0
        ELSE t.rate_value
    END AS volume_rub,
    ROW_NUMBER() OVER (PARTITION BY user_id ORDER BY operation_date) AS rn
FROM
    income AS i
LEFT JOIN
    t
ON
    t.rate_currency = i.currency
    AND DATE(t.rate_date) = DATE(i.operation_date)
"""
```

```python
pd.read_sql(sql4, connection)
```

```python
sql4_sum = """
WITH t AS
(
SELECT
    rate_currency,
    rate_countercurrency,
    CAST(CASE
        WHEN 
            LENGTH(rate_value) > 10
            THEN SUBSTR(rate_value, 6, 2) || "." || SUBSTR(rate_value, 1, 4)
        ELSE
            rate_value
    END AS REAL) as rate_value,
    rate_date
FROM
    currencies
),
tt AS
(
SELECT
    user_id,
    currency,
    operation_date,
    volume * CASE
        WHEN
            t.rate_value IS NULL
            THEN 1.0
        ELSE t.rate_value
    END AS volume_rub,
    ROW_NUMBER() OVER (PARTITION BY user_id ORDER BY operation_date) AS rn
FROM
    income AS i
LEFT JOIN
    t
ON
    t.rate_currency = i.currency
    AND DATE(t.rate_date) = DATE(i.operation_date)
)
SELECT
    SUM(volume_rub) AS sum_volume_rub
FROM
    tt
WHERE
    rn = 1
"""
```

```python
pd.read_sql(sql4_sum, connection)
```

## 5) Посчитайте по каждому клиенту среднюю разницу в днях между его пополнениями, выведите медианное значение по всем клиентам. 

```python
sql5 = """
WITH t AS
(
SELECT
    i.*,
    LAG(i.operation_date) OVER (PARTITION BY user_id ORDER BY operation_date) as prev_operation_date
FROM 
    income AS i
),
tt AS
(
SELECT
    user_id,
    AVG(JulianDay(operation_date) - JulianDay(prev_operation_date)) AS mean_delta_days
FROM
    t
WHERE
    prev_operation_date IS NOT NULL
GROUP BY
    user_id
)
SELECT
    AVG(mean_delta_days) AS median
FROM
    (
    SELECT
        mean_delta_days
    FROM
        tt
    ORDER BY
        mean_delta_days
    LIMIT 2 - (
        SELECT
            COUNT(*)
        FROM
            tt
        ) % 2
    OFFSET (
        SELECT
            (COUNT(*) - 1) / 2
        FROM tt
        )
    )
"""
```

```python
pd.read_sql(sql5, connection)
```

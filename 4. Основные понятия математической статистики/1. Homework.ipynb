Выборочное среднее арифметическое с точностью до 1 знака

```python
import numpy as np
```

```python
math_grades = [85, 90, 78, 92, 88, 85, 76, 98, 94, 87]
print(str(round(np.mean(math_grades), 1)).replace('.', ','))
```

Выборочная медиана с точностью до 1 знака

```python
print(str(round(np.median(math_grades), 1)).replace('.', ','))
```

Выборочное стандартное отклонение с точностью до 3 знаков

```python
print(str(round(np.std(math_grades, ddof=1), 3)).replace('.', ','))
```
